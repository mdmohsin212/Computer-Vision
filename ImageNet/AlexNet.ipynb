```python
import keras
from tensorflow.keras import layers
```

```python
def create_alexnet():
    """
    AlexNet architecture (2012)
    Input: 224x224x3 RGB images
    Output: 1000 classes (ImageNet)
    """
    model = keras.Sequential([
        # Conv Block 1
        layers.Conv2D(96, (11, 11), strides=4, activation='relu', input_shape=(224, 224, 3)),
        layers.MaxPooling2D((3, 3), strides=2),
        layers.BatchNormalization(),
        
        # Conv Block 2
        layers.Conv2D(256, (5, 5), activation='relu', padding='same'),
        layers.MaxPooling2D((3, 3), strides=2),
        layers.BatchNormalization(),
        
        # Conv Block 3
        layers.Conv2D(384, (3, 3), activation='relu', padding='same'),
        
        # Conv Block 4
        layers.Conv2D(384, (3, 3), activation='relu', padding='same'),
        
        # Conv Block 5
        layers.Conv2D(256, (3, 3), activation='relu', padding='same'),
        
        # Flatten
        layers.Flatten(),
        
        # Fc layers
        layers.Dense(4096, activation='relu'),
        layers.Dropout(0.5),
        layers.Dense(4096, activation='relu'),
        layers.Dropout(0.5),
        layers.Dense(1000, activation='softmax')
    ], name='AlexNet')
    
    return model
```

```python
alexnet = create_alexnet()
alexnet.summary()
```

**Comparative Study:**

```python
def simple_cnn():
    return keras.Sequential([
        layers.Conv2D(32, 3, activation='tanh', input_shape=(224, 224, 3)),
        layers.MaxPooling2D(2),
        layers.Flatten(),
        layers.Dense(1000, activation='softmax')
    ])

simple = simple_cnn()

print(f"Simple CNN parameters: {simple.count_params():,}")
print(f"AlexNet parameters: {alexnet.count_params():,}")
```
